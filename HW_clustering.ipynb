# Домашнее задание по кластеризации
### Посмотрите, как кластеризуются тексты: подберите подходящее количество кластеров, объясните, какие группы текстов выделяются в вашем датасете

# Данные

- Можно взять любой готовый датасет с текстами (min 1000 текстов), про которые заранее известна некоторая классификация (например: тема, автор, жанр и т.д.)
> Например, выберите датасет на сайте https://www.kaggle.com/datasets
(можно вбить в поиск "text classification" и "text clustering").
- Можно создать датасет самостоятельно: обкачать новости или паблики в соцсетях (если вы умеете), взять несколько книжек (в txt) и поделить по главам.


Если вы не хотите ничего искать и выбирать, возмите датасет с описанием книг [book_descriptions.csv](https://github.com/tbkazakova/compling_for_lyceum/blob/main/data/book_descriptions.csv). Кластеризируйте по аннотациям из столбца `ann`.


# Задание
1. предобработайте данные и векторизируйте тексты (CountVectorizer или TfidfVectorizer)
2. кластеризуйте (например, AgglomerativeClustering, KMeans, можно взять и другой алгоритм)
3. поясните выбранное количество кластеров и визуализируйте их облаками слов
4. примените понижение размерности (TSNE или PCA) - расположите точки текстов на плоскости (поэкспериментируйте с параметрами, если с дефолтными визуализация выглядит как туча повышенной плотности: она должна быть разреженной)
5. раскрасьте точки в зависимости от кластера, раскрасьте точки в зависимости от заранее известного класса - прокомментируйте различия и сходства визуализаций

Если вы взяли датасет с описанием книг, заранее известный класс - рекомендованный возраст (столбец `age`).
