````python
# маленький текст легко прочитать, возьмём Л. Н. Толстого
with open('vojna-i-mir.txt', encoding='utf-8') as f:
    book = f.read()
# print(book)  # проверить, что текст на месте
````

````python
print(book[7050:7150])
````

````python
tokens_with_boundaries = list(tokenize(book))  # получили список токенов с границами
# tokens_with_boundaries[:10] # посмотреть первые 10
tokens = [_.text for _ in tokens_with_boundaries]
tokens[550:560]
````

````python
tokens = []
for word in tokens_with_boundaries:
    tokens.append(word.text)
tokens[550:560]
````

````python
"4ого".isalpha()  # проверяет, что в строке только буквы
````

````python
"ЭтО нЕ лЕсЕнКа.".lower()
````

````python
# нужно убрать знаки препинания (.isalpha()), привести к нижнему регистру (.lower())
words = [w.lower() for w in tokens if w.isalpha()]
words[3600:3610]
````

````python
words = []
for w in tokens:
   if w.isalpha():
       words.append(w.lower())
words[3600:3610]
````

Итак, у нас есть список слов с маленькой буквы.

Посмотрим частотные?

````python
print(counter)  # Не очень информативно. В чём дело?
````

- надо убрать частотные, незначимые
- надо заменить слова на начальные формы

````python
# вернёмся к списку частотных чуть позже

parsed_words = []
for w in words:
    parsed_words.append(morph.parse(w)[0])
````

````python
parsed_words[3600:3602]
````

````python
# сохраним только нужное
data = []
for word in parsed_words:
    razbor = {'lex': word.normal_form,
              'word': word.word,
              'POS': word.tag.POS,  # POS - часть речи, Part of speech
              'Род': word.tag.gender,
              'Одуш': word.tag.animacy
              }
    data.append(razbor)
````

## Нарисуем графики

**Линейный график**

(знаете про закон Ципфа?)

````python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
````

````python
plt.figure(figsize=(12, 6))
plt.title('Частотность слов') # название
fdist = FreqDist(lex_list)
fdist.plot(30, cumulative=False)

plt.show()
````

**Закон Ципфа**: если все слова языка (или просто достаточно длинного текста) упорядочить по убыванию частотности их использования, то частотность n-го слова в таком списке окажется приблизительно обратно пропорциональной его порядковому номеру n (так называемому рангу этого слова, см. шкала порядка). Например, второе по используемости слово встречается примерно в два раза реже, чем первое, третье — в три раза реже, чем первое, и так далее.

(Работает не только в лингвистике > закон скорее про статистику и устройство мира)

````python
# Теперь без стопслов
plt.figure(figsize=(12, 6))
fdist = FreqDist(lex_notstop)
fdist.plot(30,cumulative=False)
plt.show()
````
