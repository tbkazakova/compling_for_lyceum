В чём разница между `.loc()` и `.iloc()`?

```python
df.index = ['Mark', 'Jane', 'Shura']
```

```python
df
```

```python
df.iloc[0]
```

```python
# df.loc[0]
df.loc['Mark']
```

Итого: `.iloc()` - для числовых индексов, `.loc()` - для индексов таблицы, выделенных жирным
